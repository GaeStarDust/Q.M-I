###Example
We can use Latex on a colab notebook (in markdown cells)
$$\psi(x,t)=e^{ikx - w\omega t}$$

The spatial part
$$\psi(x)=e^{ikx} = \cos(kx) + i sin(kx)$$

The real part:
$$\rm I\!R(\psi(x)) = \cos(kx)$$

Where k is the wavenumber(1/lenght)

```python
#Vectors for the X and Y axes

#Use spacing between point in the vector
x = np.arange(-3*np.pi, +3*np.pi,0.01)

#Define the wavenumber/Spactial frequency
k1 = 1
k2 = 2.5
k3 = 5.0
#Evaliute the Y axis
y1 = np.cos(k1*x)
y2 = np.cos(k2*x)
y3 = np.cos(k3*x)
```

## 4. 1D Plotting

```python
#Plotting Environment
plt.figure(figsize=(7,3))
#Plot customisation
plt.plot(x, y1, color = "red", linestyle= "-.", label = r"$y_1 = \sin(k_1*x)$")
plt.plot(x, y2, color = "blue", linestyle= "-", label = r"$y_1 = \sin(k_1*x)$")
plt.plot(x, y3, color = "green", linestyle= "dotted", label = r"$y_1 = \sin(k_1*x)$")
#Add the legend
plt.legend(loc=1)
#Customize the axes labels
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

#Add axes limits.
plt.xlim(-3*np.pi, +3*np.pi)
plt.ylim(-1.1, 1.5)
plt.show()
```

### 5. 2D PLotting

Now we want to represent a 2D particle:
$$\psi(x,y)= cos(k_x x)+sin(k_y y)$$
Where $k_x$ and $k_y$ are the wavenumber along X and Y(1/lenght)



```python
#To represnt 2D surfaces, we need a meshgrid

#Create 1D vectors for the axis

x = np.linspace(-3*np.pi, +3*np.pi,500)
y = np.linspace(-3*np.pi, +3*np.pi,500)

#Create the 2D meshgrid
x_2d, y_2d = np.meshgrid(x, y)

```

```python
#Python Function
def psi_2d(x_2d, y_2d):
  """
  This is what we call a docstring (what the function will do)
  Inputs: x_2d, y_2d
  Outputs: psi_2d
  """

  psi_2d = np.cos(k_x*x_2d) + np.sin(k_y*y_2d)

  return psi_2d
```

```python
#Define the wavenumber
k_x = 1.
k_y = 1.
#We call the function
psi_xy = psi_2d(x_2d,y_2d)

print(psi_xy.shape)
```

```python
#Plotting: Figure Enviroment
fig = plt.figure(figsize=(8,8))
#Axis Object
ax = fig.add_subplot(111, projection=  "3d")
#Plot
#surf = ax

surf = ax.plot_surface(x_2d, y_2d, psi_xy, cmap = "inferno")
#Add labels
ax.set_xlabel(r"$<x$")
ax.set_ylabel(r"$y$")
ax.set_zlabel(r"$z$")
#This may help

#Show Plot
plt.show(surf)
```
